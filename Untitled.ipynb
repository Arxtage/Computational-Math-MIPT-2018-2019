```python
#include <stdio.h>
#include <cmath>
#include <stdlib.h>

const int imax = 100, jmax = 100;
double u[imax+1],u1[imax+1],u2[imax+1],an[imax+1],h,tau,t,eps,max,c;
int i;

void progonka();

int main()
{ 

FILE *file; 
file = fopen("lab6_100.dat","w");

h=1.0/imax;
tau=1.0/jmax;
t=0;
eps = 0.0001;

for(i=0; i<=imax; i++)
{
u[i]= pow(h*i, 4.0/3.0)/pow(81.0, 1.0/3.0);
u1[i]= pow(h*i, 4.0/3.0)/pow(81.0, 1.0/3.0);
fprintf(file, "%lf %lf %lf\n",t, h*i, u[i]);
}



do
{
t = t + tau;

do
{
progonka();
max = 0;
for(i=0; i<=imax; i++)
{
if(fabs((u2[i]-u1[i])/u2[i])>=max)
{
max = fabs((u2[i]-u1[i])/u2[i]);
}
}

for(i=0; i<=imax; i++)
{
u1[i]=u2[i];
}


}while(max>eps);


for(i=0; i<=imax; i++)
{
an[i]=pow((h*i), 4.0/3.0)/pow((9-26*t/3.0), 2.0/3.0);
fprintf(file, "%lf %lf %lf %lf\n",t, h*i, u2[i],an[i]);
u[i]=u2[i];
u1[i]=u2[i];
}

}while (t<=1.0);

return 0;
}

void progonka()
{
double x,f1,f2, a[3], b[3], c[3], d[3], alf[imax+1], bet[imax+1];
int count;


a[0] = 0;
b[0] = 1;
c[0] = 0;
d[0] = 0;

a[2] = 0;
b[2] = 1;
c[2] = 0;
d[2] =1.0/pow(9-26*t/3, 2.0/3.0);
count = 0;

alf[0] = -a[0]/b[0];
bet[0] = d[0]/b[0];

for(i=1; i<=imax; i++) 
{
x = i*h; 

f1 = (x+h/2)*(x+h/2)*(pow((u1[i+1]),1.5)+pow((u1[i]),1.5))/(2*x*x*h*h);
f2 = (x-h/2)*(x-h/2)*(pow((u1[i]),1.5)+pow((u1[i-1]),1.5))/(2*x*x*h*h);

a[1] = -f1;
b[1] = 1.0/tau + f2 + f1;
c[1] = -f2;
d[1] = u[i]/tau;
/* if ((fabs(b[1])) < (fabs(a[1])+fabs(c[1])))
{
count++;
}*/
alf[i] = -a[1]/(b[1]+c[1]*alf[i-1]);
bet[i] = (d[1]-c[1]*bet[i-1])/(b[1]+c[1]*alf[i-1]);
}

/* if (count==0)
{
printf("ustoychiv");
}*/

u2[imax] = (d[2]-c[2]*bet[imax-1])/(b[2]+c[2]*alf[imax-1]);


for(i=imax-1; i>=0; i--) 
{
u2[i]=(alf[i])*(u2[i+1])+bet[i];

}

printf("%.8lf %lf\n",alf[imax],bet[imax]);

}
```
