```python
import matplotlib.pylab as plt
import numpy as np
get_ipython().magic('matplotlib inline')
```

```python
a = 1
b = 2
eps = 1e-04
m = 10
m2 = 20
```

```python
def F(x,y):
    return ((4*x+1)*y - 4*x - y**2)/(x*(2*x-1))
```

```python
def RK(n):
    h = float(b-a) / n
    yr = []
    x = np.linspace(a , b ,n + 1) 
    y = np.zeros(n + 1)
    f = np.zeros(4)
    y [0] = 2
    yr.append(y[0])
    for i in range(n):
        f[0] = F(x[i], y[i])
        f[1] = F(x[i] + 2*h / 3, y[i] + 2*h * f[0] / 3)
        f[2] = F(x[i] + 2*h / 3, y[i] - h*f[0]/3 + h*f[1])
        y[i+1] = y[i] + h * (f[0] + 2*f[1] +f[2]) / 4
        yr.append(y[i + 1])
    return (yr[n-1], x , yr)
```

```python
print RK(10)
```

```python
while(1):
    if (((abs((RK(m)[0]) - RK(m2)[0]))/(2**3 - 1)) < eps):
        break
    else:
        m  *= 2

        m2 *= 2
print(m)
print(m2)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.125, 0.547727, 1, 1])

ax1.yaxis.set_label_position('right')

ax1.xaxis.set_label_position('bottom')

ax2 = fig.add_axes([1.25, 0.547727, 1, 1]) 

ax2.yaxis.set_label_position('right')

ax2.xaxis.set_label_position('bottom')

for ax in fig.axes:
    ax.grid(True)

ax1.plot(RK(m)[1], RK(m)[2], 'g', RK(m)[1], RK(m)[2], 'rv')
ax2.plot(RK(m2)[1], RK(m2)[2], 'g', RK(m2)[1], RK(m2)[2], 'rv')
```
